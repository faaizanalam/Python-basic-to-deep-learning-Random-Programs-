```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
dt = pd.read_csv("student_loan_data.csv")
dt
```

```python
X = pd.DataFrame(dt.drop(['loan_status', 'loan_intent','person_education','person_home_ownership','person_gender',
             'previous_loan_defaults_on_file'], axis = 1))
y = pd.Series(dt['loan_status'])
X
```

```python
print(len(X))
len(y)
```

```python
model = GaussianNB()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= 0.8, random_state=42)
```
