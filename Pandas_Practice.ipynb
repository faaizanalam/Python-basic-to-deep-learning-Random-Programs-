### Data Visulalization using Pandas

```python
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Sales': [100, 120, 150, 200, 180]}
dt = pd.DataFrame(data)
dt
dt.plot(x='Year',y='Sales',title = 'Sales over time')
```

```python
# Create a DataFrame
data = {'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
        'Population': [8398748, 3980408, 2716000, 2326006]}
dt = pd.DataFrame(data)
dt.plot(x = 'City', y= 'Population', kind ='bar', title = 'Population by City')
```
