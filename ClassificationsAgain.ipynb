```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import r2_score, accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
```

```python
X, y = make_classification(
    n_features = 7,
    n_classes = 4,
    n_samples = 1200,
    random_state = 3,
    n_informative = 3,
    n_clusters_per_class = 1
)
```

```python
plt.scatter(X[:, 0], X[:, 1], c =y, marker= '+')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size = 0.8, shuffle = True, random_state = 42
)
```

```python
model = KNeighborsClassifier()
```

```python
#model = svm.SVC()
```

```python
#model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
print("Trying to predict a single value using the trained model")
y_pred = model.predict([X_test[4]])
print("Actual Value : ",y_test[4])
print("Predicted Value : ",y_pred[0])
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy Score",accuracy_score(y_test,y_pred))
print("Accuracy Score",accuracy_score(y_pred,y_test))
print("F1 Score", f1_score(y_test,y_pred, average = 'weighted'))
print("R2 Score", r2_score(y_test,y_pred))
```

```python
labels = [0,1,2,3]
cm = confusion_matrix(y_test, y_pred, labels = labels)
print(cm)
con = ConfusionMatrixDisplay(confusion_matrix= cm, display_labels= labels)
con.plot();
```
