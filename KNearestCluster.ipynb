```python
from sklearn.cluster import KMeans
import numpy as np 
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
data = load_digits()
X = data.data 
```

```python
kmeans = KMeans(n_clusters=10, random_state = 42, n_init=10)
```

```python
kmeans.fit(X)
```

```python
Kcenters = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
print("Centeroids : ",Kcenters)
```

```python
print("labels: ",labels)
```

```python
pred= kmeans.predict(X)
len(pred)
```

```python
sum(pred == 1)
```

```python
for i in range(10):
    print(sum(pred == i))
```

```python
fig, axes = plt.subplots(1, 10, figsize=(10, 3))
for ax, cluster_center, cluster_label in zip(axes, Kcenters, range(10)):
    ax.set_title(f"Cluster {cluster_label}")
    ax.imshow(cluster_center.reshape(8, 8), cmap="viridis")
    ax.axis("off")
plt.show()
```

```python
# Plot the clusters
plt.figure(figsize=(10, 10))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', s=30, alpha=0.7)
plt.colorbar(label='Cluster')
plt.xlabel(f"Feature {0} (Pixel Intensity)")
plt.ylabel(f"Feature {1} (Pixel Intensity)")
plt.title("Clusters of Digits Using Two Selected Features")
plt.show()
```

```python
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(data.data)

# Plot the clusters
plt.figure(figsize=(10, 8))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=kmeans.labels_, cmap='viridis', s=30, alpha=0.7)
plt.colorbar(label='Cluster')
plt.title("Clusters of Digits (PCA Reduced to 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
